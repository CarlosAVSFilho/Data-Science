```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from yellowbrick.classifier import ConfusionMatrix
```

```python
# A biblioteca keras utiliza o TensorFlow como backend.
# Obs.: A versão do TensorFlow instalada é a 1.5 pois meu processador não tem suporte a AVX. As versões do TF acima da 1.5
# necessitam deste suporte.

from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
base = datasets.load_iris()

previsores = base.data
classe = base.target
```

```python
classe_dummy = np_utils.to_categorical(classe) #A camada de saída terá 3 neurônios
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores,
                                                                  classe_dummy,
                                                                  test_size = 0.3,
                                                                  random_state = 0)
```

```python
modelo = Sequential() #Modelo com uma camada após a outra
modelo.add(Dense(units = 5, input_dim = 4)) # Camada oculta com 5 neurônios. input_dim = qnts atributos/qnts neuronios terão
# na camada de entrada. (4 no caso, dados da pétala e sépala)
modelo.add(Dense(units = 4))
modelo.add(Dense(units = 3, activation = 'softmax')) #Camada de saída. Softmax: problemas de classific com mais de 2 classes

modelo.summary()
```

```python
# Imprimir
from keras.utils.vis_utils import plot_model
```

```python
# Compilando a rede
modelo.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
modelo.fit(X_treinamento, y_treinamento, epochs = 1000, validation_data = (X_teste, y_teste))
```

```python
previsoes = modelo.predict(X_teste)
```

```python
previsoes = (previsoes > 0.5)
```

```python
# v = ConfusionMatrix(modelo) # N é possível gerar a matriz de confusão com o yellow brick para estes casos

y_teste_matrix = [np.argmax(t) for t in y_teste]
y_previsao_matrix = [np.argmax(t) for t in previsoes]
```

```python
confusao = confusion_matrix(y_teste_matrix, y_previsao_matrix)
confusao
```
